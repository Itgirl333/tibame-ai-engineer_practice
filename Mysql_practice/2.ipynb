```python
import pymysql
import prettytable
import IPython.display as dp

host = input('請輸入主機位置:')
user = input('請輸入資料庫帳號:')
password = input('請輸入資料庫密碼:')
port = int(input('請輸入資料庫的Port:'))
database = input('請輸入資料庫的名稱:')

con = pymysql.connect(host = host,
                    user = user,
                    password = password,
                    port = port,
                    database = database)

cur = con.cursor()
dp.clear_output(True)
text = """(0) 離開程式
(1) 顯示會員列表
(2) 新增會員資料
(3) 更新會員資料
(4) 刪除會員資料
"""

def showall():
    table = prettytable.PrettyTable(['編號', '姓名', '生日', '地址'], encoding = 'utf-8')
    cur.execute("SELECT * FROM member")
    data = cur.fetchall()
    for d in data:
        table.add_row([d[0], d[1], d[2], d[3]])
    table.align['名稱'] = 'l'
    dp.clear_output(True)
    print(table)

while True:
    print(text)
    command = input('指令:')
    if command == '1':
        showall()
    elif command == '2':
        dp.clear_output(True)
        name = input('請輸入會員姓名:')
        birthday = input('請輸入會員生日:')
        address = input('請輸入會員地址:')
        sql = "INSERT INTO member (name, birthday, address) VALUES (%s, %s, %s);"
        cur.execute(sql, [name, birthday, address])
        con.commit()
        showall()
    elif command == '3':
        dp.clear_output(True)
        idx = input('請選擇你要修改的資料編號:')
        name = input('請輸入會員姓名:')
        birthday = input('請輸入會員生日:')
        address = input('請輸入會員地址:')
        sql = "UPDATE member SET name=%s,birthday=%s,address=%s where id=%s;"
        cur.execute(sql, [name, birthday, address, idx])
        con.commit()
        showall()
    elif command == '4':
        dp.clear_output(True)
        idx = input('請選擇你要刪除的資料編號:')
        sql = "DELETE FROM member WHERE id=%s"
        cur.execute(sql, (idx))
        showall()
    elif command == '0':
        cur.close()
        con.close()
        break
```
