```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
img = Image.open('bike.jpeg')
img = np.array(img)
filter_x = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
filter_y = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
img_x = cv2.filter2D(img, -1, filter_x)
img_y = cv2.filter2D(img, -1, filter_y)
img_ = cv2.addWeighted(img_x, 0.5, img_y, 0.5, 0)
img_total = np.hstack([img_x, img_y, img_])
plt.figure(figsize=(15,5))
plt.imshow(img_total, cmap='gray')
plt.axis('off')
```
